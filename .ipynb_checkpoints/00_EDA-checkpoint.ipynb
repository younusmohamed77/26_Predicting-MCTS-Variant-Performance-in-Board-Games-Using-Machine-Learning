- Entries: 233,234 rows and 814 columns, indicating a large and high-dimensional dataset.
- Numerical Columns: 809 columns (with 201 as float64 and 608 as int64), covering game and agent properties.
- Categorical Columns: 5 object-type columns, likely including descriptive information such as game names and agent identifiers.
- Properties, Format, Time, Discrete, Realtime: These columns are mostly binary or constant values (1.0 or 0.0), suggesting they represent static game properties that may not vary much across rows.
- num_wins_agent1, num_draws_agent1, num_losses_agent1 columns provide counts of outcomes.
- Target Variable (utility_agent1):
    - Mean: 0.044 (slightly positive bias)
    - Range: [-1.0, 1.0], with quartiles showing:
        - 25% of values below -0.33 (indicating losses)
        - 50% around 0 (draws)
        - 75% above 0.53 (indicating wins)
- Performance Metrics (PlayoutsPerSecond, MovesPerSecond) - Wide range, especially in MovesPerSecond (from 5.69 to over 2 million), possibly reflecting different game complexities.

### Key Insights
#### Feature Engineering Opportunities:
- Constant and Binary Columns: Columns with little to no variance can be dropped or consolidated, reducing dimensionality without sacrificing information.
- Outcome and Performance Metrics: Counts and per-second metrics may help model the effectiveness of various MCTS configurations.
- Target Variable Analysis: With values centered around 0, the model should handle balanced win-loss scenarios and adapt to slight bias in agent advantage.
- These insights provide guidance on potential feature reduction and relevant metrics for a predictive model.

### 2. Missing Values

```python
# Check for missing values
missing_values = train_data.isnull().sum()
print("\nMissing Values in train dataset:")
print(missing_values[missing_values > 0])
```

```python
# Check for missing values
missing_values_test = test_data.isnull().sum()
print("\nMissing Values in test dataset:")
print(missing_values_test[missing_values_test > 0])
```

- In train and data, 18 columns are completely missing (233,234 and 3 missing values), covering features like Behaviour, StateRepetition, Duration, etc.
- Since these columns have no values in train and test data, they do not contribute useful information for model training. Dropping them reduces unnecessary data load and simplifies the model.
- Removing these columns in both datasets ensures a uniform feature set across train and test data, facilitating smooth model training and prediction without handling mismatched columns.

```python
# Target variable distribution (utility_agent1)
plt.figure(figsize=(10, 5))
sns.histplot(train_data['utility_agent1'], kde=True, bins=30)
plt.title('Distribution of Utility (utility_agent1)')
plt.xlabel('Utility')
plt.ylabel('Frequency')
plt.show()
```

```python
# Unique count of categorical columns
print("\nUnique values in categorical columns:")
categorical_cols = ['GameRulesetName', 'agent1', 'agent2']
for col in categorical_cols:
    print(f"{col}: {train_data[col].nunique()} unique values")
```

```python
# Agent1 and Agent2 Variants - Breakdown
agent1_breakdown = train_data['agent1'].value_counts().head(10)
agent2_breakdown = train_data['agent2'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=agent1_breakdown.values, y=agent1_breakdown.index, palette='viridis')
plt.title('Top 10 Agent1 Variants')
plt.xlabel('Count')
plt.ylabel('Agent1 Variant')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=agent2_breakdown.values, y=agent2_breakdown.index, palette='viridis')
plt.title('Top 10 Agent2 Variants')
plt.xlabel('Count')
plt.ylabel('Agent2 Variant')
plt.show()
```

```python
# Correlation heatmap for numerical features
numerical_cols = train_data.select_dtypes(include=np.number).columns.tolist()
plt.figure(figsize=(12, 10))
correlation_matrix = train_data[numerical_cols].corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap for Numerical Features')
plt.show()
```

```python
# Feature engineering - break down agent columns for analysis
train_data['agent1_selection'], train_data['agent1_exploration'], train_data['agent1_playout'], train_data['agent1_score_bounds'] = zip(
    *train_data['agent1'].str.split('-').apply(lambda x: x[1:5])
)
train_data['agent2_selection'], train_data['agent2_exploration'], train_data['agent2_playout'], train_data['agent2_score_bounds'] = zip(
    *train_data['agent2'].str.split('-').apply(lambda x: x[1:5])
)
```

```python
# Pair plot for agent1 components with utility_agent1
sns.pairplot(train_data[['utility_agent1', 'agent1_selection', 'agent1_exploration', 'agent1_playout', 'agent1_score_bounds']], 
             hue='agent1_selection')
plt.suptitle('Pair Plot of Agent1 Components and Utility', y=1.02)
plt.show()
```

```python
# Countplot of selection strategy distribution for Agent1 and Agent2
plt.figure(figsize=(10, 5))
sns.countplot(x='agent1_selection', data=train_data, palette='viridis', order=train_data['agent1_selection'].value_counts().index)
plt.title('Agent1 Selection Strategy Distribution')
plt.xlabel('Selection Strategy')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='agent2_selection', data=train_data, palette='viridis', order=train_data['agent2_selection'].value_counts().index)
plt.title('Agent2 Selection Strategy Distribution')
plt.xlabel('Selection Strategy')
plt.ylabel('Count')
plt.show()
```
